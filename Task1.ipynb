```python
import re
import os

def register_user():
    username = input('Enter Username: ')
    if (retrieve_password(username)=='not_found'):
        if('@' in username and '.' in username):
            a=username.split('@')
            if(len(a)==2 and a[0] != '' and a[1][0]!='.' and \
            bool(re.search('^[a-zA-Z]', a[0])) == True \
            and a[1][0]!='.'):
                print("Username is valid")
            else:
                print('Invalid Username. Please provide a valid one..')
                return
            password = input('Enter Password: ')
            pswd_pattern = '^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[!@#$&*%?])[a-zA-Z\d!@#$&*%?]{5,16}$'
            result = re.findall(pswd_pattern,password)
            if (result):
                print('Valid Password')
                register_valid_user(username, password)
            else:
                print('Invalid Password')
        else:
            print('Invalid Username. Please provide a valid one..')
    else:
        print('Username already exists. Try other username..')
    return

def register_valid_user(username, password):
    f = open('credentials.txt', 'a')
    f.write(username + ',' + password + '\n')
    f.close()
    print('Registration Successful!')

def login():
    username = input('Enter Username: ')
    password = input('Enter Password: ')
    f_pswd = retrieve_password(username)
    # print(f_pswd)
    f_pswd = f_pswd.strip()
    if(password == f_pswd and password != 'not_found'):
        print('Successfully Logged in')
    elif(f_pswd == 'not_found'):
        print('User not found. Please Register')
    else:
        print('Incorrect Password')
    return

def retrieve_password(username):
    try:
        f = open('credentials.txt', 'r')
    except FileNotFoundError as e:
        print('You are the first user..')
        return 'not_found'
    lines = f.readlines()
    for line in lines:
        if(len(line)>1):
            f_user, f_pswd = line.split(',')
            if (f_user == username):
                f.close()
                return f_pswd
    f.close()
    return 'not_found'

def reset_pswd(username):
    if (retrieve_password(username)!='not found'):
        password = input('Enter New Password: ')
        hai = '^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[!@#$&*%?])[a-zA-Z\d!@#$&*%?]{5,16}$'
        result = re.findall(hai, password)
        if result:
            temp_pswd_file=open('temp_pswd_file.txt','a')
            f = open('credentials.txt', 'r')
            lines = f.readlines()
            for line in lines:
                if(len(line)>1):
                    f_user,  f_pswd= line.split(',')
                    if  (username!=f_user):
                        temp_pswd_file.write(f_user + ',' + f_pswd)
                    else:
                        temp_pswd_file.write(f_user + ',' + password + '\n')
            f.close()
            temp_pswd_file.close()
            os.remove('credentials.txt')
            os.rename('temp_pswd_file.txt','credentials.txt')
            print('Password changed successfully')




        else:
            print('Password not valid. Please retry.')
    else:
        print('Not a valid user. Please register.')


while(True):
    print("1. Registration\n2. Login \n3. Forgot Password\n4. Exit\n")
    option = int(input('Enter your choice: '))

    if(option == 4):
        print('Thank you.. Bye!')
        exit()
    elif(option == 1):
        register_user()

    elif(option == 2):
        login()
    elif(option==3):
        username = input('Enter username : ')
        reset_pswd(username)
```
